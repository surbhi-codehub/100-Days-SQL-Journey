```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('/content/company.db')

# Helper function to run SQL queries
def run_sql(query):
    return pd.read_sql_query(query, conn)

print("✅ Connected to company.db successfully!")
```

```python
#1 Display all products taht are NOT in the 'Electronic' category.
run_sql("""SELECT *
FROM products
WHERE NOT category = 'Electronic';""")
```

```python
#2 Display customer who are NOT from 'Bangalore'.
run_sql("""SELECT *
FROM customers
WHERE NOT city = 'Bangalore';""")
```

```python
# 3 Display employees whose department_id is NOT 1.
run_sql("""SELECT *
FROM employees
WHERE NOT department_id = 1;""")
```

```python
#4 Display orders that were NOT placed by customer_id 1.
run_sql("""SELECT *
FROM orders
WHERE NOT customer_id = 1;""")
```

```python
#5 Display employee_id ,first_name, and last_name for employees whose department_id is NOT 2.
run_sql("""SELECT employee_id, first_name, last_name
FROM employees
WHERE NOT department_id = 2;""")
```
