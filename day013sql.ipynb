```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('/content/company.db')

# Helper function to run SQL queries
def run_sql(query):
    return pd.read_sql_query(query, conn)

print("✅ Connected to company.db successfully!")
```

```python
#1 Display customers whose city is 'Bangalore' or 'Mumbai'.
run_sql("""SELECT *
FROM customers
WHERE city = 'Bangalore' OR city = 'Mumbai'""")
```

```python
#2 Display product whose category is 'Electronic' OR 'Furniture'.
run_sql("""SELECT *
FROM products
WHERE category = 'Electronic' OR category = 'Funiture'""")
```

```python
#3 Dispaly employees whose department_id is 1 OR 2.
run_sql("""SELECT *
FROM employees
WHERE department_id = 1 OR department_id = 2;""")
```

```python
#4 Display order placed by customer_id 1 OR customer_id 2.
run_sql("""SELECT *
FROM orders
WHERE customer_id =1 OR customer_id = 2;""")
```

```python
#5 Display first_name, last_name, and email of employees whose department_id is 2 OR employee_id is greater than 8.
run_sql("""SELECT first_name, last_name, email
FROM employees
WHERE department_id = 2 OR employee_id > 8;""")
```
