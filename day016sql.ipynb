```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('/content/company.db')

# Helper function to run SQL queries
def run_sql(query):
    return pd.read_sql_query(query, conn)

print("✅ Connected to company.db successfully!")
```

```python
# Topic: Using BETWEEN to filter a range of values.
```

```python
#1 Display products whose price is BETWEEN 1000 AND 5000.
run_sql("""SELECT *
FROM products
WHERE price BETWEEN 1000 AND 5000;""")
```

```python
#2 Display employees whose employee_id is Between 3 AND 8.
run_sql("""SELECT *
FROM employees
WHERE employee_id BETWEEN 3 AND 8;""")
```

```python
#3 Display customers whose customer_id is BETWEEN 2 AND 6.
run_sql("""SELECT *
FROM customers
WHERE customer_id BETWEEN 2 AND 6;""")
```

```python
#4 Display orders whose order_id is BETWEEN 5 AND 15.
run_sql("""SELECT *
FROM orders
WHERE order_id BETWEEN 5 AND 15;""")
```

```python
#5 Display employee_id, first_name, and last_name for employees whose department_id is 2 and employee_id is BETWEEN 2 AND 10.
run_sql("""SELECT employee_id, first_name, last_name
FROM employees
WHERE department_id = 2
AND employee_id BETWEEN 2 AND 10;""")
```
