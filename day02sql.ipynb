```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('/content/company.db')

# Helper function to run SQL queries
def run_sql(query):
    return pd.read_sql_query(query, conn)

print("✅ Connected to company.db successfully!")
```

```python
run_sql("""SELECT department_name
FROM departments;""")
```

```python
run_sql("""SELECT Product_name, price
FROM products;""")
```

```python
run_sql("""SELECT DISTINCT city
FROM customers;""")
```

```python
run_sql("""SELECT DISTINCT job_title
FROM employees;""")
```

```python
run_sql("""SELECT first_name, last_name, email
FROM employees;""")
```
