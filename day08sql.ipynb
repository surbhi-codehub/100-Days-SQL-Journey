```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('/content/company.db')

# Helper function to run SQL queries
def run_sql(query):
    return pd.read_sql_query(query, conn)

print("✅ Connected to company.db successfully!")
```

```python
#1
run_sql("""SELECT employee_id, email
FROM employees
LIMIT 5;""")
```

```python
#2
run_sql("""SELECT department_name
FROM departments
LIMIT 4;""")
```

```python
#3
run_sql("""SELECT product_id, product_name, category
FROM products
LIMIT 6;""")
```

```python
#4
run_sql("""SELECT first_name, email, city
FROM customers
LIMIT 8;""")
```

```python
#5
run_sql("""SELECT order_id, customer_id, order_date
FROM orders
LIMIT 12;""")
```

```python
# Bounus Challenge
run_sql("""SELECT product_name, price
FROM products""")
```
