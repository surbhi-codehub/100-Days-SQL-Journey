```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('/content/company.db')

# Helper function to run SQL queries
def run_sql(query):
    return pd.read_sql_query(query, conn)

print("✅ Connected to company.db successfully!")
```

```python
#1 Display unique department names from teh department table.
run_sql("""SELECT DISTINCT department_name
FROM departments;""")
```

```python
#2 Display unique city values from the customers table.
run_sql("""SELECT DISTINCT city
FROM customers;""")
```

```python
#3 Dispaly unique categery values from the products table.
run_sql("""SELECT DISTINCT category
FROM products;""")
```

```python
#4 Display the first 5 unique employee job titles using DISTINCT and LIMIT.
run_sql("""SELECT DISTINCT job_title
FROM employees
LIMIT 5;""")
```

```python
#5 Display teh first 10unique customer cities usin DISTINCT and LIMIT.
run_sql("""SELECT DISTINCT city
FROM customers
LIMIT 10;""")
```

```python
# Bonus Challege
run_sql("""SELECT DISTINCT phone
FROM customers;""")
```
