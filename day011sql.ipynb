```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('/content/company.db')

# Helper function to run SQL queries
def run_sql(query):
    return pd.read_sql_query(query, conn)

print("✅ Connected to company.db successfully!")
```

```python
#1 Display all detalis of the employee whose employee_id is 1.
run_sql("""SELECT *
FROM employees
WHERE employee_id = 1;""")
```

```python
#2 Display all products whose category is 'Electronics'.
run_sql("""SELECT *
FROM products
WHERE category = 'Electronics';""")
```

```python
#3 Display all customers who belong to the belong to city 'Bangalore'.
run_sql("""SELECT *
FROM customers
WHERE city = 'Bangalore';""")
```

```python
#4 Display all orders placed by customer_id 1.
run_sql("""SELECT *
FROM orders
WHERE customer_id = 1;""")
```

```python
#5 Display first_name, last_name, and email of employees who belong to department_id 2.
run_sql("""SELECT first_name, last_name, email
FROM employees
WHERE department_id = 2;""")
```
